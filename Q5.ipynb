```python
import json
import pandas as pd
import re
import operator
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn import svm
from sklearn import linear_model
from scipy.stats.stats import pearsonr
from sklearn.metrics import mean_squared_error
```

```python
final_df = pd.read_csv('final_df_clean.csv')
```

```python
final_df
```

```python
authors_data=pd.read_csv('authors_pubs.csv')
authors_data = authors_data.merge(final_df[['Arxiv_key','h_index','m_index','e_index','i10_index','g_index','tori_index','riq_index']],on=['Arxiv_key'],left_index= False)
authors_data = authors_data.dropna().reset_index(drop = True)
```

```python
authors_data
```

```python
authors_data['author_count']=authors_data['Author'].apply(lambda x: len(x.strip('[]').split(',')))
```

```python
publications = authors_data["Pub"].unique()
small_pubs = []
big_pubs = {}

for item in publications:
    papers = authors_data[authors_data["Pub"] == item]
    data = np.array(papers["Citation_count"])
    median = np.median(data)
    if len(papers.index) < 35:
        small_pubs.append(median)
    if len(papers.index) >= 35:
        big_pubs[item]= median

publication_imp = []
small_pubs_median = np.median(small_pubs)

for index in range(len(authors_data.index)):
    pub_name = authors_data[authors_data.index == index]['Pub'].iloc[0]
    if pub_name in big_pubs.keys():
        publication_imp.append(big_pubs[pub_name])
    else:
        publication_imp.append(small_pubs_median)
```

```python
authors_data['pub_imp'] = publication_imp
```

```python
authors_data = authors_data[authors_data['pub_imp'] > 5]
```

```python
authors_data.columns
```

```python
features = ['Figures', 'Pages',
       'Year', 'Pubs(Avg.)', 'Pubs(Max)', 'Pubs(Min)', 'h_index',
       'm_index', 'e_index', 'i10_index', 'g_index', 'tori_index', 'riq_index',
       'author_count', 'pub_imp']
```

```python
x = authors_data[features]
y = authors_data.Citation_count
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 0)
```

```python
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X_train, y_train)
```

```python
regressor.feature_importances_

clf_coefs = []
for i in range(len(features)):
    clf_coefs.append([i,regressor.feature_importances_[i]])
clf_coefs = np.array(clf_coefs)
for j in range(len(clf_coefs)):
    print('%20s\t%.5f' % (features[int(clf_coefs[j,0])],clf_coefs[j,1]))
```

```python
answer = regressor.predict(x)
```

```python
indexes = list(y.index.values)
```

```python
predictions = {}
for i in range(len(answer)):
    predictions[indexes[i]] = answer[i]
```
